```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('puf2022.csv')

# Display the first 5 rows (the "head")
print(df.head())
```

```python
# Replace 'I' with 1 and 'R' with 2 in the 'jstatus' column
df["jstatus"].replace("I", 1, inplace=True)
df["jstatus"].replace("R", 2, inplace=True)

df["jprice"].replace("I", 1, inplace=True)
df["jprice"].replace("R", 2, inplace=True)
df["jprice"].replace("9", 3, inplace=True)

df["jsqft"].replace("I", 1, inplace=True)
df["jsqft"].replace("R", 2, inplace=True)
df["jsqft"].replace("9", 3, inplace=True)

df["jbedroom"].replace("I", 1, inplace=True)
df["jbedroom"].replace("R", 2, inplace=True)
df["jbedroom"].replace("9", 3, inplace=True)

df["jlease"].replace("I", 1, inplace=True)
df["jlease"].replace("R", 2, inplace=True)
df["jlease"].replace("9", 3, inplace=True)

df["jfinaldest"].replace("I", 1, inplace=True)
df["jfinaldest"].replace("R", 2, inplace=True)
df["jfinaldest"].replace("9", 3, inplace=True)

df["jtitle"].replace("I", 1, inplace=True)
df["jtitle"].replace("R", 2, inplace=True)
df["jtitle"].replace("9", 3, inplace=True)

df["jsecured"].replace("I", 1, inplace=True)
df["jsecured"].replace("R", 2, inplace=True)
df["jsecured"].replace("9", 3, inplace=True)

df["jlocation"].replace("I", 1, inplace=True)
df["jlocation"].replace("R", 2, inplace=True)
df["jlocation"].replace("9", 3, inplace=True)

df["jfootings"].replace("I", 1, inplace=True)
df["jfootings"].replace("R", 2, inplace=True)
df["jfootings"].replace("9", 3, inplace=True)


df["jpiers"].replace("I", 1, inplace=True)
df["jpiers"].replace("R", 2, inplace=True)
df["jpiers"].replace("9", 3, inplace=True)
```

```python
print(df.head())
```

```python
corr = df.corr()["PRICE"]
print(corr)
```

```python
# Create a histogram
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
sns.histplot(data=df, x="PRICE", kde=True, bins=30)  # Adjust bins as needed
plt.xlabel("PRICE")
plt.ylabel("Frequency")
plt.title("Distribution of PRICE in pur2022.csv")
plt.show()

# Create a box plot
sns.boxplot(x="PRICE", data=df, showmeans=True)
plt.xlabel("PRICE")
plt.title("Box Plot of PRICE Distribution")
plt.show()
```

```python
# Create the KDE plot
sns.kdeplot(data=df, x="PRICE", fill=True, linewidth=0)

# Customize the plot
plt.xlabel("PRICE")
plt.ylabel("Density")
plt.title("Kernel Density Estimation of PRICE Distribution")
plt.show()
```

```python
# Scatter plot for the target variable "PRICE"
plt.scatter(df.index, df['PRICE'])
plt.title('Scatter Plot for "PRICE"')
plt.xlabel('Index')
plt.ylabel('PRICE')
plt.show()
```
