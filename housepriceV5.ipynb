```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Load the data from the CSV file
data = pd.read_csv('puf2022_MinMaxScaler.csv')

# Split the data into predictors (X) and target (y)
X = data.drop(columns=['PRICE'])
y = data['PRICE']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a new random forest classifier
rfc = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model to the training data
rfc.fit(X_train, y_train)

# Predict the target variable for the testing data
y_pred = rfc.predict(X_test)

# Print the accuracy score of the model
print(f'The accuracy score of the model is: {rfc.score(X_test, y_test)}')
```

```python
import pandas as pd
from sklearn.linear_model import Lasso
from xgboost import XGBRegressor
from sklearn.ensemble import VotingRegressor
from sklearn.model_selection import train_test_split

# Load the data from the CSV file
data = pd.read_csv('puf2022_MinMaxScaler.csv')

# Split the data into predictors (X) and target (y)
X = data.drop(columns=['PRICE'])
y = data['PRICE']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a new Lasso regressor
lasso = Lasso(alpha=0.1)

# Create a new XGBoost regressor
xgb = XGBRegressor(n_estimators=100, random_state=42)

# Create a new voting regressor with 65% Lasso and 35% XGBoost
hybrid = VotingRegressor(estimators=[('lasso', lasso), ('xgb', xgb)], weights=[0.65, 0.35])

# Fit the model to the training data
hybrid.fit(X_train, y_train)

# Predict the target variable for the testing data
y_pred = hybrid.predict(X_test)

# Print the R-squared score of the model
print(f'The R-squared score of the model is: {hybrid.score(X_test, y_test)}')
```

```python
import pandas as pd
from sklearn.linear_model import Lasso
from xgboost import XGBRegressor
from sklearn.ensemble import StackingRegressor
from sklearn.model_selection import train_test_split

# Load the data from the CSV file
data = pd.read_csv('puf2022_MinMaxScaler.csv')

# Split the data into predictors (X) and target (y)
X = data.drop(columns=['PRICE'])
y = data['PRICE']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a new Lasso regressor
lasso = Lasso(alpha=0.1)

# Create a new XGBoost regressor
xgb = XGBRegressor(n_estimators=100, random_state=42)

# Create a new stacking regressor with Lasso and XGBoost
stacked = StackingRegressor(estimators=[('lasso', lasso), ('xgb', xgb)])

# Fit the model to the training data
stacked.fit(X_train, y_train)

# Predict the target variable for the testing data
y_pred = stacked.predict(X_test)

# Print the R-squared score of the model
print(f'The R-squared score of the model is: {stacked.score(X_test, y_test)}')
```
