```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("puf2022_processed.csv")

# Select only integer columns
int_cols = df.select_dtypes(include=['int64'])

# Convert integer columns to float type
df[int_cols.columns] = int_cols.astype('float64')

```

```python
# Get column names
column_names = df.columns.tolist()

# Get column types
column_types = df.dtypes.to_dict()
# Print column names and types
for col_name, col_type in zip(column_names, column_types.values()):
    print(f"Column name: {col_name}, Column type: {col_type}")
```

```python

# Check for null values
null_values = df.isnull().sum()
print(f'The number of null values in each column are:\n{null_values}\n')

# Check for duplicate rows
duplicate_rows = df.duplicated()
print(f'The number of duplicate rows in the data is: {duplicate_rows.sum()}\n')

# Remove null values
df.dropna(inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

```

```python
# Define the columns to keep
keep_cols = ['PRICE', 'WEIGHT', 'SQFT', 'FINALDEST', 'LEASE', 'BEDROOMS',
             'STATUS', 'REGION', 'TITLED', 'SECURED', 'LOCATION', 'PIERS']

# Select only the desired columns
df = df[keep_cols]

```

```python
df.shape
```

```python
import matplotlib.pyplot as plt

# Columns to visualize
cols_to_visualize = ['PRICE', 'WEIGHT', 'SQFT', 'FINALDEST', 'LEASE',
                     'BEDROOMS', 'STATUS', 'REGION', 'TITLED', 'SECURED',
                     'LOCATION', 'PIERS']

# Loop through each column and create visualizations
for col in cols_to_visualize:
    # Frequency
    print(f"Frequency for {col}:\n{df[col].value_counts()}")

    # Histogram
    plt.figure(figsize=(8, 5))
    plt.hist(df[col])
    plt.title(f"Histogram for {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

    # Boxplot
    plt.figure(figsize=(8, 5))
    plt.boxplot(df[col])
    plt.title(f"Boxplot for {col}")
    plt.xlabel(col)
    plt.show()

    # Violinplot
    plt.figure(figsize=(8, 5))
    plt.violinplot(df[col])
    plt.title(f"Violinplot for {col}")
    plt.xlabel(col)
    plt.show()
```

```python
# Save the modified DataFrame back to the CSV file
df.to_csv("puf2022_modified.csv", index=False)  # Don't save the index
```
